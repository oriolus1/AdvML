```python
def preprocessing(dataframe, country, start_date):
    df_country = df.loc[(df['location']==country) & (df['date'] >= start_date)][['date', 'total_cases', 'new_cases']]
    df_country.reset_index(drop=True, inplace=True)
    df_country.loc[df_country['new_cases']==0, 'new_cases'] = 1
    df_country_train = df_country.loc[0:49,]
    df_country_test = df_country.loc[50:,]
    return df_country_train, df_country_test
```

```python
x = df_russia_train.index.values.reshape(-1, 1)
x2 = x ** 2
const = np.ones(N).reshape(-1, 1)
Xsq = np.hstack((x2, x, const))
Xsq.shape
Xsq[:, 1]
```

```python
ln_new_cases = np.log(df_russia_train['new_cases'])
```

```python
sigmoid_model = LinearRegression(fit_intercept=False).fit(Xsq, ln_new_cases)
print(sigmoid_model.coef_)
#print(cur_model.intercept_)
plt.plot(Xsq[:, 1], sigmoid_model.predict(Xsq), linewidth=2)#, label="$d=%d$" % d)
plt.scatter(Xsq[:, 1], ln_new_cases)
plt.show()


#plt.plot(x, np.exp(cur_model.predict(x)), linewidth=2)#, label="$d=%d$" % d)
#plt.scatter(x, y)
#plt.show()
```

```python
sigma2MLsigmoid = np.sum((ln_new_cases - sigmoid_model.predict(Xsq)) ** 2) / N
sigma2MLsigmoid
```

```python
mu0 = np.zeros(3)
#print(mu0)
alpha = 1
Sigma0 = alpha * np.identity(3)

muN, SigmaN = calculate_mu_and_sigma(mu0, Sigma0, sigma2MLsigmoid, Xsq, ln_new_cases)
print(muN)
print(SigmaN)
```

```python
x_total = df_russia.index.values.reshape(-1, 1)
x_sq_total = x_total ** 2
const_total = np.ones(x_total.size).reshape(-1, 1)
X_total = np.hstack((x_sq_total, x_total, const_total))
print(X_total.shape)
y_total = np.array(df_russia['total_cases'])

plt.figure(figsize=(12, 6), dpi=80)
plt.plot(x_total, y_total, color='r', linewidth=2)

res = []

for i in range(0, 100):
    #print("i = %d" %i)
    w_sampled = np.random.multivariate_normal(muN, SigmaN)
    #print(w_sampled)
    t_sigmoid_pred_ln_new_cases = w_sampled @ X_total.T
    #print(np.around(t_sigmoid_pred_ln_new_cases[0:100], 1))
    t_sigmoid_pred_new_cases = np.exp(t_sigmoid_pred_ln_new_cases)
    t_sigmoid_pred_total_cases = np.cumsum(t_sigmoid_pred_new_cases)
    res.append(t_sigmoid_pred_total_cases)
    #print(np.around(total_cases_pred[0:100], 1))
    plt.ylim(bottom=0, top=3 * 10**6)
    plt.xlim(left=0, right=300)
    plt.plot(x_total, t_sigmoid_pred_total_cases, color='k', linewidth=0.1)
    
plt.plot(x_total, np.mean(res, axis=0), color='blue', linewidth=1)
```
